```python
import os
import numpy as np
import pandas as pd
from PIL import Image
```

```python
ruta_imagenes = './FaceData/'
lado = 30
size = (lado, lado)
```

```python
os.listdir(ruta_imagenes)
```

```python
datos = []

for nombre_archivo in os.listdir(ruta_imagenes):
    ruta_imagen = os.path.join(ruta_imagenes, nombre_archivo) # definimos la ruta de la imagen
    imagen = Image.open(ruta_imagen).convert('L') # aplicamos una escala de grises
    imagen = imagen.resize(size) # redimensionamos la imagen
    sexo = int(nombre_archivo.split('_')[1]) # obtenemos el sexo de la imagen
    vector_imagen = np.array(imagen).flatten() # convertimos la imagen a un vector file
    fila = np.append(vector_imagen, np.array([sexo])) # formamos un vector fila que contenga la imagen y el sexo
    datos.append(fila) # agregamos las imagenes a la lista de datos
```

```python
df = pd.DataFrame(datos)
df.head()
```

```python
df.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Y = df[900] # variable a predecir
X = df.drop(columns=[90]) # matriz de atributos
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test) # Se hace la inferencia sobre el conjunto de validacion
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_test, Y_pred)
```

* $C_{0,0}$ son los verdaderos negativos (caras de hombre clasificadas correctamente)
* $C_{1,1}$ son los verdaderos positivos (caras de mujer clasificadas correctamente)
