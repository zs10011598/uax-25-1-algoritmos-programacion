```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df = df.drop(columns=['PassengerId', 'Name', 'Ticket'])
df.head()
```

```python
df = df.drop(columns=['Cabin'])
df.head()
```

```python
def cod_sex(sex):
    if sex  == 'female':
        return 1
    elif sex=='male':
        return 2
```

```python
df['Sex'] = df['Sex'].apply(lambda x: cod_sex(x))
df.head()
```

```python
df['Embarked'].unique()
```

```python
df['Embarked'] = df['Embarked'].fillna('X')
df['Embarked'].unique()
```

```python
def cod_emb(x, e):
    if x == e:
        return 1
    else:
        return 0
```

```python
df['Embarked_S'] = df['Embarked'].apply(lambda x: cod_emb(x, 'S'))
df['Embarked_C'] = df['Embarked'].apply(lambda x: cod_emb(x, 'C'))
df['Embarked_Q'] = df['Embarked'].apply(lambda x: cod_emb(x, 'Q'))
df['Embarked_X'] = df['Embarked'].apply(lambda x: cod_emb(x, 'X'))
df.head()
```

```python
df = df.drop(columns=['Embarked'])
df.head()
```

```python
df.columns
```

```python
df = df.dropna(subset=['Age'])
df.shape
```

```python
df[df['SibSp'].isna()].shape[0]
```

```python
y_train = df['Survived']
X_train = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
       'Embarked_S', 'Embarked_C', 'Embarked_Q', 'Embarked_X']]
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
dv = dv.drop(columns=['Embarked'])
dv.head()
```

```python
dv = dv.dropna(subset=['Age'])
dv.shape
```

```python
dv = dv.dropna(subset=['Fare'])
dv.shape
```

```python
X_test = dv[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
       'Embarked_S', 'Embarked_C', 'Embarked_Q', 'Embarked_X']]
```

```python
y_pred = knn.predict(X_test)
```
